```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('tips.csv')
```

```python
df
```

```python
# Line Chart - Tip amount vs Day of the week
plt.figure(figsize=(10, 6))
sns.lineplot(x='day', y='tip', data=df)
plt.title('Tip Amount by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Tip Amount')
plt.grid(True)
plt.show()
```

```python
# Barplot - Tip amount by Sex
plt.figure(figsize=(8, 6))
sns.barplot(x='sex', y='tip', data=df)
plt.title('Tip Amount by Sex')
plt.xlabel('Sex')
plt.ylabel('Tip Amount')
plt.grid(True)
plt.show()
```

```python
# Heatmap - Correlation matrix (excluding categorical features)
correlation = df.select_dtypes(include=[np.number]).corr()  # Filter for numerical columns
plt.figure(figsize=(8, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Scatterplot - Tip amount vs Total bill amount
plt.figure(figsize=(10, 6))
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.title('Tip Amount vs Total Bill Amount')
plt.xlabel('Total Bill Amount')
plt.ylabel('Tip Amount')
plt.grid(True)
plt.show()
```

```python
# Histogram - Distribution of total bill amount
plt.figure(figsize=(8, 6))
sns.histplot(df['total_bill'])
plt.title('Distribution of Total Bill Amount')
plt.xlabel('Total Bill Amount')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
# Boxplot - Tip amount by Smoker
plt.figure(figsize=(8, 6))
sns.boxplot(x='smoker', y='tip', data=df)
plt.title('Tip Amount by Smoker')
plt.xlabel('Smoker (Yes or No)')
plt.ylabel('Tip Amount')
plt.grid(True)
plt.show()
```

```python
# Violin Plot - Tip amount by Party size
plt.figure(figsize=(8, 6))
sns.violinplot(x='size', y='tip', data=df)
plt.title('Tip Amount by Party Size')
plt.xlabel('Party Size')
plt.ylabel('Tip Amount')
plt.grid(True)
plt.show()
```

```python
# Time Series Charts - Tips by Day
plt.figure(figsize=(14, 6))
sns.lineplot(x='day', y='tip', data=df, hue='time')
plt.xlabel('Day')
plt.ylabel('Tip')
plt.grid(True)
plt.show()
```
