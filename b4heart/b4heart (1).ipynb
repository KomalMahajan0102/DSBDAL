```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('heart.csv')
```

```python
df
```

```python
# Line Chart - Example: Blood Pressure vs. Age (choose numeric features for line charts)
plt.figure(figsize=(10, 6))
sns.lineplot(x='age', y='trestbps', data=df)  # Replace with appropriate numeric features
plt.title('Resting Blood Pressure by Age')
plt.xlabel('Age')
plt.ylabel('Resting Blood Pressure (mmHg)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.title('Bar Chart')
plt.xlabel('Age')
plt.ylabel('Chol')
plt.bar(df['age'], df['chol'], color='red')
plt.show()
```

```python
# Scatterplot - Explore relationships between numeric features (e.g., age vs max heart rate)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='thalach', data=df)  # Replace with relevant features
plt.title('Maximum Heart Rate by Age')
plt.xlabel('Age')
plt.ylabel('Maximum Heart Rate')
plt.grid(True)
plt.show()
```

```python
# Histogram - Distribution of a numeric variable (e.g., fasting blood sugar)
plt.figure(figsize=(8, 6))
sns.histplot(df['fbs'])
plt.title('Distribution of Fasting Blood Sugar')
plt.xlabel('Fasting Blood Sugar (1 = > 120 mg/dL)')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
# Boxplot - Distribution of a numeric variable by a categorical variable (e.g., cholesterol by sex)
plt.figure(figsize=(8, 6))
sns.boxplot(x='sex', y='chol', data=df)
plt.title('Cholesterol Levels by Sex')
plt.xlabel('Sex (0 = Female, 1 = Male)')
plt.ylabel('Cholesterol (mg/dL)')
plt.grid(True)
plt.show()
```

```python
# Violin Plot - Similar to boxplot but shows kernel density estimation
plt.figure(figsize=(8, 6))
sns.violinplot(x='sex', y='thalach', data=df)
plt.title('Distribution of Maximum Heart Rate by Sex')
plt.xlabel('Sex (0 = Female, 1 = Male)')
plt.ylabel('Maximum Heart Rate')
plt.grid(True)
plt.show()
```

```python
# Heatmap - Correlation Matrix
plt.figure(figsize=(8, 8))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
